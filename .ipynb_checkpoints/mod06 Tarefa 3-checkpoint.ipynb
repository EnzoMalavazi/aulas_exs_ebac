#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline 
%matplotlib notebook
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5']>= 8) & (df['APGAR5']<= 10),'nivel'] = 'normal'
df.loc[(df['APGAR5']>= 6) & (df['APGAR5']<= 7),'nivel'] ='Asfixia_leve'
df.loc[(df['APGAR5']>= 4) & (df['APGAR5']<= 5),'nivel'] ='Asfixia_moderada'
df.loc[(df['APGAR5']>= 0) & (df['APGAR5']<= 3),'nivel'] ='Asfixia_severa'

df['nivel']
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
fig2 = sns.catplot(x = 'GESTACAO',
                   
                   y = 'APGAR5',
                   
                   hue = 'SEXO',
                   
                   data = df , 
                   
                   kind = 'bar',
                   
                   ci = None)

fig2.set_xticklabels(rotation=45)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
crtab = pd.crosstab(df['GESTACAO'], df['APGAR5'])
crtab
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
crtab.sum(axis='columns')/(10*100)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
fig3 = sns.catplot( data = crtab,
                   
                   kind = 'bar',
                   
                   ci = None)

fig2.set_xticklabels(rotation=45)
```
